```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('salary_and_number_of_childeren_data.csv')

y = df['Salary']/10000
x = df['Number of Children']
m = len(x)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
b = 7
w = 1.5

plt.scatter(x, y)

axes = plt.gca()
x_vals = np.array(axes.get_xlim())
y_vals = b + w * x_vals
plt.plot(x_vals, y_vals, '--', color='red')

plt.show()
```

```python
def compute_cost(x, y, w, b): 

    m = x.shape[0] 
    
    cost_sum = 0 
    for i in range(m): 
        f_wb = w * x[i] + b   
        cost = (f_wb - y[i]) ** 2  
        cost_sum = cost_sum + cost  
    total_cost = (1 / (2 * m)) * cost_sum  

    return total_cost
```

```python
compute_cost(x , y, w, b)
```

```python
w = np.linspace(0, 10, 100)
```

```python
j = []

for i in w:
    j.append(compute_cost(x , y, i, b))
             
j = np.array(j)

j
```

```python
plt.plot(w, j)
plt.show()
```
